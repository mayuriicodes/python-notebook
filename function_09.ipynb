**FUNCTION WITH DEFAULT ARGUMENTS**

```python
print('function with arguments')
def tax_cal():
    salary=eval(input('enter salary:'))
    tax_per=eval(input('enter tax:'))
    tax_pay=salary*tax_per/100
    print(tax_per)
tax_cal()

print('function without arguments')
def tax_cal1(salary,tax_per):
      tax_pay=salary*tax_per/100
      print(tax_per)
tax_cal1(10000,20)    

print('function with default arguments')
def tax_cal2(salary,tax_per=10):
      tax_pay=salary*tax_per/100
      print(tax_per)
tax_cal2(50000)
```

```python
#function without argument
def even_odd():
    num=eval(input('enter a number'))
    if num%2==0 :
        print(f'{num} is even ')
    else :
        print(f'{num} is odd ')
even_odd()
```

```python
#function with argument
def even_odd1(num):
    if num%2==0 :
        print(f'{num} is even ')
    else :
        print(f'{num} is odd ')
even_odd1(10)
```

```python
#fuction with default argument
def even_odd2(num=10):
    if num%2==0 :
        print(f'{num} is even ')
    else :
        print(f'{num} is odd ')
even_odd2()
```

```python
def add(n):
    if n==0:
        return(0)
    else:
        return(n+(add(n-1)))
add(5)               
```

```python
def fact(n):
    if n==0:
        return(0)
    else:
        return(n*(fact(n-1)))
fact(10)
```
