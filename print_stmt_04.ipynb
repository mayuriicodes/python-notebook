```python
num1=1000
num2=2000
add=num1+num2
sub=num1-num2
mul=num1*num2
div=num1/num2
print("addition of {} an {} is {}".format(num1,num2,add))
print("substraction of {} an {} is {}".format(num1,num2,sub))
print("multiplication of {} an {} is {}".format(num1,num2,mul))
print("division of {} an {} is {}".format(num1,num2,div))
```
