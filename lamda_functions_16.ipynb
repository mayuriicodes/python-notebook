**lambda function**
- similar analogy to list comprehension
- represents single line
- to reduce the time complexity


**$pattern-1$ function with one argument**

```python
def mul(a):
    return(a*a)
ans=mul(10)
ans
```

```python
#syntax
#function_name=lambda <arg> : <return output>
#function_name=mul
#arg name = a
#return output =a*a
```

```python
mul=lambda a : a*a
mul(20)
```

```python
def cube(a):
    return(a*a*a)
cube(30)
```

```python
cube= lambda a : a*a*a
cube(20)
```

**$pattern-2$ function with two args**

```python
def add(a,b):
    return(a+b)
add(10,20)
```

```python
add= lambda a,b : a+b
add(10,20)
```

**$pattern-3$function with three args**

```python
def avg(a,b,c):
    return (a+b+c)/3
avg(10,20,30)
```

```python
avg= lambda a,b,c : (a+b+c)/3
avg(10,20,30)
```

```python
avg= lambda a,b,c : round(a+b+c)/3
avg(10,22,30)
```

**$pattern-4$ function with default args**

```python
def avg(a,b,c=30):
    return round(a+b+c)/3
avg(10,20)
```

```python
avg= lambda a,b,c=30 : round(a+b+c)/3
avg(10,22)
```

**$pattern-4$ **
- list cases

```python
input=['hyd','pune','mumbai','mindurbusiness']
#output=['Hyd','Pune','Mumbai','Mindurbusiness']
```

```python
output=[]
for i in input:
    output.append(i.capitalize())
output
```

```python
output=[i.capitalize() for i in input]
output
```

```python
#above is for list comprehension
```

```python
#now, for lambda for lambda function
```

```python
#step-1: lambda<variable> :<outout>
lambda i :i.capitalize()
```

```python
#step-2: lambda<variable> :<outout> <iterator>
lambda i :i.capitalize(),input
```

```python
#step-3: map(lambda<variable> :<outout> <iterator>)
map(lambda i :i.capitalize(),input)
```

```python
#step-3: list(map(lambda<variable> :<outout> <iterator>))
list(map(lambda i :i.capitalize(),input))
```

```python
list(map(lambda i : input.index(i),input))
```

```python
input='hello how arw you'
list(map(lambda i : i.capitalize(),input.split()))
```

**filter**


- filter will use instead of map

```python
input=['hyd','bang#uluru','mumbai','pu#ne']
list(filter(lambda i : '#' in i , input))
```

```python
input=['hyd','bang#uluru','mumbai','pu#ne']
list(filter(lambda i : '#' not in i , input))
```

**reduce**

- reduce is used to perform math operation using lambda function

```python
l1=[1,2,3,4,5]
summ=0
for i in l1:
    summ=summ+i
print(summ)
```

```python
summ=0
[summ:=summ+i for i in l1]
print(summ)
```

```python
from functools import reduce
reduce(lambda summ,i:summ+i,l1)
```
