**slices**

- slice means cut into pieces
- it has similar analogy like for loop range
- needs (start,stop,step)

```python
count=0
str1='ola ola ola'
for i in range(len(str1)):
    if str1[i:i+3]=='ola':  #SLICE WINDOW==== "str1[i:i+3]"===FIXED WINDOW
        count=count+1
print(count)
```

**count**

```python
count=0
str1='ola ola ola'
for i in range (len(str1)):
    if str1[i]=='a':
         count=count+1
print(count)
```
