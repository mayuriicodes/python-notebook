**list comprehension**
- list comprehension is a way to write list operations in a single line
- when we apply for loop and if-else conditional statements then the code copilation become slow
- which means it takes time 
- in order to avoid the time complexity we will use list comprehension
- list comprehension is a single line code

$only -for -loop$

```python
lst1=[]
for i in range (10):
    lst1.append(i)
lst1
```

```python
#syntax
lst1=[<code should write here>]
#inside code no need of append
lst1=[<output var> <for loop>]
```

```python
lst1=[i for i in range(10)]
lst1
```

**dont give ':' after for loop**

```python
input=['hyd','pune','mumbai','mindurbusiness']
output=[i.capitalize() for i in input]
output
```

```python
input=['hyd','pune','mumbai','mindurbusiness']
output=[]
for i in input:
    output.append(i.upper()) 
output
```

$for loop- if condition$

```python
#even number
even_list=[]
for i in range(1,10):
    if i%2==0:
        even_list.append(f'even:{i}')
even_list
```

```python
even_list=[]
[f'even:{i}' for i in range(1,10) if i%2==0]
#<output>    <for loop>          <if condition>
```

```python
input1=['hyd','pune','mum#bai','mindurbusi#ness']
output1=[]
for i in input1:
    if '#' in i:
        output1.append(i)
output1
```

```python
input1=['hyd','pune','mum#bai','mindurbusi#ness']
output1=[i for i in input1 if '#' in i]
output1
```

```python
input1=['hyd','pune','mum#bai','mindurbusi#ness']
output1=[i for i  in input1 if '#' not in i]
output1
```

**for if-else**

```python
even=[]
odd=[]
for i in range(1,10):
    if i%2==0:
        even.append(f'even:{i}')
    else:
        odd.append(f'odd: {i}')
print(even)
print(odd)
```

```python
even=[]
odd=[]
[f'even:{i}' if i%2==0 else f'odd: {i}' for i in range(1,10)]
```

**distance between two points**




```python
import math
l1=[10,20] #[x1,y1]
l2=[30,40] #[x2,y2]
d1=l2[0]-l1[0] #(x2-x1)^2  30-10=20
d1_sq=math.pow(d1,2)
d1_sq
d2=l2[1]-l1[1] #(y2-y1)^2  40-20=20
d2_sq=math.pow(d2,2)
d2_sq
summ=d1_sq+d2_sq
math.sqrt(summ)
```

```python
math.sqrt(math.pow(l2[0]-l1[0],2)+math.pow(l2[1]-l1[1],2))
```
