```python
summ=0
for i in range(1,11):
    summ=summ+i
print(summ)
```

```python
num=eval(input('enter the number which divisors you want :'))
for i in range(1,num+1):
    if num%i==0:
      num=num+i
print(num)
```

**counter wrapper**
- counting the success ones
- count=0
- for loop
    - count=count+1

```python
count=0
num=eval(input('enter the number :'))
for i in range(1,num+1):
    if num%i==0:
      print(i)
      count=count+1
print('the number of divisors are:' ,count)
```

```python
import random
even_count,odd_count=0,0
odd_sum, even_sum=0,0
for i in range(5):
    num=random.randint(1,100)
    if num%2==0:
        print(f'{num} is even')
        even_count=even_count+1
        even_sum=even_sum+i
    else:
        print(f'{num} is odd')
        odd_count=odd_count+1
        odd_sum=odd_sum+i
print('number of evens are: ',even_count)
print('number of odds are: ',odd_count)
print('sum of evens are: ',even_sum)
print('sum of odds are: ',odd_sum)

    
```
