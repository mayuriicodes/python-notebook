**zip**

```python
name=['ram','sita','shyam']
age=[20,30,25]
for i,j in zip(name,age):
    print(f'{i} age is {j}')
```

```python
name=['ram','sita','shyam']
age=[20,30,25]
[f'{i} age is {j}' for i,j in zip(name,age) ]
```

**dictionary comprehension**

```python
#syntax
{<key:value <for loop>}
```

```python
names=['ramesh','sush','ram']
age=[20,30,25]
{i:j for i,j in zip (names,age)}
```
