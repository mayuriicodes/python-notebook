$if-else$

```python
import random
n=random.randint(-100,100)
if n>0:
    print(f'{n} is a positive number')
else:
    print(f'{n} is a negative number')

```

```python
n=eval(input('enter the number'))
if n>0:
    print('number is a positive number')
else:
    print('number is a negative number')
```

```python
n1=eval(input('enter the num1 :'))
n2=eval(input('enter the num2 :'))
if n2>n1:
    print(f'{n2} is greater')
else:
    print(f'{n1} is greater')
```

```python
import random
n1=random.randint(0,100)
n2=random.randint(0,100)
if n2>n1:
    print(f'{n2} is greater')
else:
    print(f'{n1} is greater')
```

```python
distance=eval(input('enter the distance :'))
if distance>25:
    charge=eval(input('enter the charge per km :'))
    total_charge=distance*charge
    print(f' the total charges is {total_charge} INR')
    
else :
    print('the ride is free')
    
    
```

```python
free_distance=25
distance=eval(input('enter the distance :'))
if distance>25:
    charge=eval(input('enter the charge per km :'))
    ACTUAL_distance=distance-free_distance
    total_charge=ACTUAL_distance*charge
    print(f'the actual distance covered is {ACTUAL_distance}')
    print(f' the total charges is {total_charge} INR')
    
else :
    print('the ride is free')
    
```

```python
import random
n1=eval(input("enter your number between 1 to 10 :"))
n2=random.randint(1,10)
#print(f'my number is :{n2}')
if n2==n1:
    print('you won')
else:
    print('you lost')
```

$if-elif-else$

```python
#syntax
if <con>:
    <write your code here >
elif <con>:
    <write your code here >
elif <con>:
    <write your code here >
else :
    <write your code here >
```

```python
n=eval(input('enter num :'))
if n==1:
        print('one')
elif n==2:
        print('two')
elif n==3:
        print('three')
else :
    print('enter a valid number')

      
```
