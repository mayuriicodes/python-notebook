#type casting
- convert one data type into another data type

**int to other datatypes**

```python
num= 100
type(num)
```

```python
float(num)
```

```python
bool_num=bool(num)
type(bool_num)
```

```python
#quick check
float(100), bool(100), str(100)
```

**float to another datatype**

```python
flo=10.5
type(flo)
```

```python
int(flo)
```

```python
print(int(10.5))
print(str(10.5))
print(bool(10.5))
```
