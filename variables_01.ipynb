# variables

- variables are used to store data
- data can be change 
- these variables can be used to manipulate python code

```python
num=100
#100 is stored in variable called num
num
```

**rules for variables**

- variable should begin with underscore or alphabet.
    - ex: _city ='pune'
            num = 10
- should not begin with number or any special symbol.
    - ex: @city = 'pune' 
           1num = 10    
- keywords cannot be variable 
    - ex: if = 10 

```python
n1 =10   #10 is stored in 'n1' variable
n2 =20   #20 is stored in 'n2' variable
n3 =n1+n2  #addition of n1 and n2 will store in variable 'n3'
n3
```
