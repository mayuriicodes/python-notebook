**TRY-EXCEPT**

- TRY AND EXCEPT IS A CONCEPT OF EXCEPTION HANDLING
- TWO BLOCKS
   - 1.TRY BLOCK
   - 2. EXCEPT BLOCK
- ORIGINAL CODE WILL RUN UNDER TRY BLOCK
- IF ANY ERROR OCCURS IT'LL  CATCH BY EXCEPT BLOCK

```python
try:
    a=10
    b=0
    a/b
except:
    print("error is in try block therefor i'm printing")
    
```

- if the code is correct except block will not execute.
- except block is execute only when there is an error in try block.

```python
try:
    n1=eval(input('enter num1 :'))
    n2=eval(input('enter num2 :'))
    div=n1/n2
    print('the div is' , div)
except Exception as e :
    print(e)
    

```
